```python
def input_rename(i, noi):
    noi_name = '{0:01}'.format(i)+'{0:02}'.format(noi)
    
    # gets the current working dir
    os.chdir(str(DIR) +"\input runs")
    src_dir = os.getcwd()
    
    src_dir_1=src_dir + "\D Par FlexInvOptionUB_"+str(noi_name)+".inc"
    src_dir_2=src_dir + "\D Set FlexInvOptionRun_"+str(noi_name)+".inc"
    
    dst_dir_1=str(DIR) +"\input" + "\D Par FlexInvOptionUB"+".inc"
    dst_dir_2=str(DIR) +"\input" + "\D Set FlexInvOptionRun"+".inc"


    shutil.copy(src_dir_1, dst_dir_1)
    shutil.copy(src_dir_2, dst_dir_2)
input_rename (0,1)
```
